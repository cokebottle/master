1.1 Să se cifreze textul reprezentat de numele dumneavoastră de familie, prin algoritmul Cezar. Cheia utilizata este numărul de litere din numele de familie. Verificaţi rezultatul prin descifrare.

Nume: __CIOBANU__

Cheie: __7__

Operatie: __ROT(7)__

Alphabet | ABCDEFGHIJKLMNOPQRSTUVXYZ
     ---|---
Cipher  | HIJKLMNOPQRSTUVWXZABCDEFG
Text criptat = JPVIHUB

```python
nume = "ciobanu"
key = 7
def enc(str):
    encrypted = []
    for litera in str:
        encrypted.append(chr((ord(litera) - ord('a') + key) % 26 + ord('a')))
    result = "".join(encrypted)
    return result

def dec(str):
    decrypted = []
    for litera in str:
        decrypted.append(chr((ord(litera) - ord('a') - key) % 26 + ord('a')))
    result = "".join(decrypted)
    return result

if __name__== "__main__":
    print(enc(nume))
    print(dec(enc(nume)))
```
